```python
import pandas as pd
import numpy as np
import datetime
```

```python
d={
    'hash'          :'category',
    'trajectory_id' :'category',
    'time_entry'    :'category',
    'time_exit'    :'category',
    'vmax':'float64',
    'vmin':'float64',
    'vmean':'float64',
    'x_entry':'float64',
    'y_entry':'float64',
    'x_exit':'float64',
    'y_exit':'float64'
}
```

```python
train = pd.read_csv('./input/data_train.csv',dtype=d).drop(columns=['Unnamed: 0'])
test = pd.read_csv('./input/data_test.csv',dtype=d).drop(columns=['Unnamed: 0'])
train['time_entry']=train['time_entry'].astype('datetime64[ns]')
trajectory_id = test['trajectory_id']
```

```python
def within_measure(x, y):
    #  3750901.5068 ≤ 𝑥 ≤ 3770901.5068
    #  −19268905.6133 ≤ 𝑦 ≤ −19208905.6133
    if 3750901.5068 <= x and x <= 3770901.5068 and -19268905.6133 <= y and y <= -19208905.6133:
        return 1
    else:
        return 0

def time_to_sec(values):
    return pd.to_timedelta(values).dt.total_seconds().astype(int)

```

```python
X = train['x_exit'].values 
Y = train['y_exit'].values
city = [within_measure(x, y) for x,y in zip(X,Y)]

```

```python
test2 = test[(test['time_exit']>datetime.timedelta(15,0,0))&(test['time_exit']<datetime.timedelta(16,0,0))]
```

```python
train.time_entry = time_to_sec(train.time_entry)
train.time_exit = time_to_sec(train.time_exit)
test.time_entry = time_to_sec(test.time_entry)
test.time_exit = time_to_sec(test.time_exit)



```

```python
for col in test.columns:
    test[col].fillna(test[col].mode()[0], inplace=True)
    train[col].fillna(train[col].mode()[0], inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

X_train = train.drop(columns=['hash','trajectory_id'])
X_test = test.drop(columns=['hash','trajectory_id'])

model = RandomForestClassifier(n_estimators=100, max_depth=2,
                             random_state=0)
model.fit(X_train, city)


pred = model.predict(X_test)
```

```python
submission= pd.DataFrame()
submission['id']=trajectory_id 
submission['target'] = pred
submission.to_csv('./output/submission.csv',index=False)    
submission[:20]
```
